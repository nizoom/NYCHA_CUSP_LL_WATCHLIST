```python
import requests
import pandas as pd
import os
from dotenv import load_dotenv

load_dotenv()
app_token = os.getenv("TOKEN_KEY")

url = "https://data.cityofnewyork.us/api/v3/views/wvxf-dwi5/query.json"
all_rows = []
page = 1
page_size = 50000  # max out page size to minimize API calls

while True:
    params = {
        "pageNumber": page,
        "pageSize": page_size,
        "app_token": app_token
    }
    response = requests.get(url, params=params)
    response.raise_for_status()
    
    rows = response.json() # adjust key based on what you saw earlier
    all_rows.extend(rows)
    print(f"Page {page} — total rows so far: {len(all_rows)}")
    
    if len(rows) < page_size:  # last page
        break
    page += 1

df_large = pd.DataFrame(all_rows)
df_large.to_parquet("large_dataset.parquet", index=False)
print("Saved.")
```
