# Confirming overlap between NYCHA violation data and General building violation data 

```python
import pandas as pd
from dotenv import load_dotenv
import os
load_dotenv()
```

```python
nycha_df = pd.read_csv("../data/nycha.csv")
app_token = os.getenv("TOKEN_KEY")
```

### Getting NYCHA BINS

```python
nycha_df['BIN'].shape
```

```python
nycha_bins = nycha_df['BIN'].dropna().unique().tolist()
nycha_bins = [int(x) for x in nycha_bins]
len(nycha_bins)
```

### Creating params

```python
id_list = ", ".join(f"'{b}'" for b in nycha_bins)
```

```python
hpd_violations = pd.read_csv('../data/Housing_Maintenance_Code_Violations.csv')
```

```python
hpd_violations.columns
```

```python
hpd_w_NYCHA_bins = hpd_violations[hpd_violations['BIN'].isin(nycha_bins)]
```

```python
print(hpd_w_NYCHA_bins.shape)
len(hpd_w_NYCHA_bins['BIN'].unique())
```

```python
print(nycha_df['BIN'].dtype)
print(hpd_violations['BIN'].dtype)
```

```python
hpd_violations['BIN'].isnull().sum()

hpd_violations.shape
```

```python
print(nycha_df.columns.tolist())
print(nycha_df.shape)
```

```python
# BIN in nycha is float64 (e.g. 1234567.0), convert to nullable int then string
nycha_df["BIN_key"] = nycha_df["BIN"].astype("Int64").astype(str)

# BIN in large dataset is object — strip any whitespace/decimals just in case
hpd_violations["BIN_key"] = hpd_violations["BIN"].astype("Int64").astype(str).str.strip()

# check they look the same now
print(nycha_df["BIN_key"].head(10).tolist())
print(hpd_violations["BIN_key"].head(10).tolist())
```

```python
small_bins = set(nycha_df["BIN_key"].dropna())
large_bins = set(hpd_violations["BIN_key"].dropna())

matched_bins = small_bins & large_bins
print(f"NYCHA BINs:         {len(small_bins)}")
print(f"Matched in large:   {len(matched_bins)}")
print(f"Match rate:         {len(matched_bins)/len(small_bins)*100:.1f}%")
```
